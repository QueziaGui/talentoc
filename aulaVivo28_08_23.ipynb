```python
#Estoque
estoque = {'Toalha_Banho': 15, 'Toalha_Rosto': 20, 'Roupão_fem': 10, 'Roupão_Masc': 25, 'Roupão_infantil_f':30, 'Roupão_inf_m': 10}

#Mostrar lista de produtos e quantidade
def mostrar_estoque():
  print("Lista de produtos e quantidades")
  for produto, quant in estoque.itens():
    print(f"- {produto}: {quant}")

#Mudar nome do produto
def mudar_nome_produto():
  print("Digite o produto que deseja trocar o nome: ")
  produto_anterior = input()
  if produto_anterior in estoque:
    print(f"Quantos {produto_anterior} você que trocar? ")
    quant = int(input())
    if quant <= estoque[produto_anterior]:
      print(f"Digite o nome do novo produto: ")
      prod_novo = input()
      print(f"Digite a quantidade do novo produto: ")
      quant_novo = int(input)
      estoque[produto_anterior] -= quant_novo
      if estoque[produto_anterior] == 0:
        del estoque[produto_anterior]
      if prod_novo in estoque:
        estoque[prod_novo] += quant_novo
      else:
        estoque[prod_novo] = quant_novo
      print(f"Você modificou {quant} {produto_anterior} por {quant_novo} {prod_novo}.")
    else:
      print(f"Você não possui{quant} {produto_anterior} no estoque.")
  else:
    print(F"Você não possui {produto_anterior}na lista.")

#Adicionar produtos

def add_produto():
  print("Digite o nome do produto que será incluído: ")
  produto = input()
  print(f"Digite a quantidade do produto inseridoincluído: ")
  quant = int(input())
  if produto in estoque:
    estoque[produto] += quant
  else:
    estoque[produto] = quant
  print(f"Você adicionou {quant} {produto}à lista.")

#Menu
print("===============CONTROLE DE ESTOQUE===============")
print("Escolha uma opção:")
print("-------------------------------------------------")
print("1 - Mostrar os produtos e sua quantidade ")
print("-------------------------------------------------")
print("2 - Mudar um produto por outro ")
print("-------------------------------------------------")
print("3 - Adicionar mais produtos ao sistema ")
print("-------------------------------------------------")
print("4 - Sair! ")

opcao =  int(input())

while opcao != 4:
  if opcao == 1:
    mostrar_estoque()
  elif opcao == 2:
    mudar_nome_produto()
  elif opcao ==3:
    add_produto()
  else:
    print("Opçao inválida")

  print("Escolha novamente!")
  opcao = int(input())

print("Agradecemos sua participação!")




```
